```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('image.jpg')
```

```python
# 
# 
blur = cv2.blur(img, (3, 3))
plt.imshow(blur[:,:,::-1])
plt.show()
```

```python
# 
# normalizeTrue
# normalizeFalse
blur = cv2.boxFilter(img, -1, (3, 3), normalize=True)
plt.imshow(blur[:,:,::-1])
plt.show()
blur = cv2.boxFilter(img, -1, (2, 2), normalize=False)
plt.imshow(blur[:,:,::-1])
plt.show()
```

```python
# 
# 
gaussian = cv2.GaussianBlur(img, (5, 5), 1)
plt.imshow(gaussian[:,:,::-1])
plt.show()
```

```python
# 
# 
median = cv2.medianBlur(img, 5)
plt.imshow(median[:,:,::-1])
plt.show()
```

```python
# 
# 
blur = cv2.bilateralFilter(img, 9, 75, 75)
plt.imshow(blur[:,:,::-1])
plt.show()
```
