```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('image.jpg')
```

```python
# 固定值填充
# 固定值边框，统一都填充0也称为zero padding
# src：要处理的原图
# top, bottom, left, right：上下左右要扩展的像素数
# borderType：边框类型，这个就是需要关注的填充方式
# cv2.BORDER_CONSTANT这种方式就是边框都填充成一个固定的值
cons = cv2.copyMakeBorder(img, 10, 5, 10, 5, cv2.BORDER_CONSTANT, value=(255, 255, 255))
```

```python
plt.imshow(cons[:,:,::-1])
plt.show()
```

```python
# 默认边框cv2.BORDER_DEFAULT其实是取镜像对称的像素填充
cons = cv2.copyMakeBorder(img, 50, 50, 50, 50, cv2.BORDER_DEFAULT)
plt.imshow(cons[:,:,::-1])
plt.show()
```

```python
# OpenCV卷积
# 定义卷积核
kernel = np.ones((3, 3), np.float32) / 10
# 卷积操作，-1表示通道数与原图相同
dst = cv2.filter2D(img, -1, kernel)
plt.imshow(np.hstack((img, dst))[:,:,::-1])
plt.show()
```
