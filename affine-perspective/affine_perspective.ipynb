```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('image.jpg')
rows, cols = img.shape[:2]
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
# 仿射变换
# 变换前的三个点
pts1 = np.float32([[50, 50], [50, 300], [300, 300]])
# 变换后的三个点
pts2 = np.float32([[50, 80], [50, 350], [300, 200]])
```

```python
M = cv2.getAffineTransform(pts1, pts2)
dst = cv2.warpAffine(img, M, (cols, rows))
```

```python
plt.subplot(121)
plt.imshow(img[:,:,::-1]) 
plt.title('input')
plt.scatter(pts1[:,0], pts1[:,1])
plt.subplot(122)
plt.imshow(dst[:,:,::-1])
plt.title('output')
plt.scatter(pts2[:,0], pts2[:,1])
plt.show()
```

```python
# 透视变换
img = cv2.imread('card.jpg')
```

```python
# 原图中卡片的四个角点
pts1 = np.float32([[148, 80], [437, 114], [94, 247], [423, 288]])
# 变换后分别在左上、右上、左下、右下四个点
pts2 = np.float32([[0, 0], [320, 0], [0, 178], [320, 178]])
# 生成透视变换矩阵
M = cv2.getPerspectiveTransform(pts1, pts2)
# 进行透视变换，参数3是目标图像大小
dst = cv2.warpPerspective(img, M, (320, 178))
```

```python
plt.subplot(121), plt.imshow(img[:, :, ::-1]), plt.title('input')
plt.subplot(122), plt.imshow(dst[:, :, ::-1]), plt.title('output')
plt.show()
```

```python
# 透视变换
img = cv2.imread('card1.jpg')
# 原图中卡片的四个角点
pts1 = np.float32([[29, 79], [260, 24], [106, 409], [339, 353]])
# 变换后分别在左上、右上、左下、右下四个点
pts2 = np.float32([[0, 0], [230, 0], [0, 300], [230, 300]])
# 生成透视变换矩阵
M = cv2.getPerspectiveTransform(pts1, pts2)
# 进行透视变换，参数3是目标图像大小
dst = cv2.warpPerspective(img, M, (230, 300))
plt.subplot(121), plt.imshow(img[:, :, ::-1]), plt.title('input')
plt.subplot(122), plt.imshow(dst[:, :, ::-1]), plt.title('output')
plt.show()
```
