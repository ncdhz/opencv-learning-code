```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('image.jpg')
```

### 亮度对比度处理公式
$$
g(x) = \alpha f(x) + \beta
$$

```python
# 亮度与对比度
res = np.uint8(np.clip((1.1 * img + 5), 0, 255))
```

```python
plt.imshow(np.hstack((img, res))[:,:,::-1])
plt.show()
```
