```python
import cv2
import matplotlib.pyplot as plt
```

```python
# 缩放图片
img = cv2.imread('image.jpg')
# 按照指定的宽度、高度缩放图片
res = cv2.resize(img, (125, 140))
# 按照比例缩放，如x,y轴均放大一倍
# 缩放方法interpolation
res2 = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_LINEAR)
plt.imshow(res[:,:,[2,1,0]])
plt.show()
plt.imshow(res2[:,:,[2,1,0]])
plt.show()
```

```python
# 参数2 = 0：垂直翻转(沿x轴)，参数2 > 0: 水平翻转(沿y轴)，参数2 < 0: 水平垂直翻转。
dst1 = cv2.flip(img, 1)
dst2 = cv2.flip(img, 0)
dst3 = cv2.flip(img, -1)
titles = ['Original', 'Flip horizontally', 'Flip vertically', 'Flip horizontally and vertically']
images = [img, dst1, dst2, dst3]
for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.imshow(images[i][:,:,[2,1,0]])
    plt.title(titles[i], fontsize=8)
    plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import numpy as np
# 平移图片
rows, cols = img.shape[:2]

# 定义平移矩阵，需要是numpy的float32类型
# x轴平移100，y轴平移50
M = np.float32([[1, 0, 100], [0, 1, 50]])
# 用仿射变换实现平移
dst = cv2.warpAffine(img, M, (cols, rows))

plt.imshow(img[:,:,[2,1,0]])
plt.show()
plt.imshow(dst[:,:,[2,1,0]])
plt.show()
```

```python
# 参数1：图片的旋转中心
# 参数2：旋转角度(正：逆时针，负：顺时针)
# 参数3：缩放比例，0.5表示缩小一半
M = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 0.5)
dst = cv2.warpAffine(img, M, (cols, rows))
plt.imshow(img[:,:,[2,1,0]])
plt.show()
plt.imshow(dst[:,:,[2,1,0]])
plt.show()
```
