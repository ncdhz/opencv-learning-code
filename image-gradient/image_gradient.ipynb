```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('sudoku.jpg', 0)
kernel = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
], dtype=np.float32)
```

```python
# 垂直边缘提取
dst_v = cv2.filter2D(img, -1, kernel)
# 水平边缘提取
dst_h = cv2.filter2D(img, -1, kernel.T)
```

```python
plt.imshow(np.hstack((img, dst_v, dst_h)), 'gray')
plt.show()
```

```python
# Sobel算子
# 参数2：ddepth: 输出图像的深度（可以理解为数据类型），-1表示与原图像相同的深度
# 参数3、4：dx,dy:当组合为dx=1,dy=0时求x方向的一阶导数，当组合为dx=0,dy=1时求y方向的一阶导数（如果同时为1，通常得不到想要的结果）
sobelx = cv2.Sobel(img, -1, 1, 0, ksize=3) # 只计算x方向
sobely = cv2.Sobel(img, -1, 0, 1, ksize=3) # 只计算y方向
```

```python
plt.imshow(np.hstack((img, sobelx, sobely)), 'gray')
plt.show()
```

```python
# Prewitt算子
kernel = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
], dtype=np.float32)
# 垂直边缘提取
dst_v = cv2.filter2D(img, -1, kernel)
# 水平边缘提取
dst_h = cv2.filter2D(img, -1, kernel.T)
plt.imshow(np.hstack((img, dst_v, dst_h)), 'gray')
plt.show()
```

```python
# Scharr算子
kernel = np.array([
    [-3, 0, 3],
    [-10, 0, 10],
    [-3, 0, 3]
], dtype=np.float32)
# 垂直边缘提取
dst_v = cv2.filter2D(img, -1, kernel)
# 水平边缘提取
dst_h = cv2.filter2D(img, -1, kernel.T)
plt.imshow(np.hstack((img, dst_v, dst_h)), 'gray')
plt.show()
```

```python
# Laplacian算子
laplacian = cv2.Laplacian(img, -1)
plt.imshow(np.hstack((img, laplacian)), 'gray')
plt.show()
```

```python
kernel = np.array([
    [1, 1, 1],
    [1, -8, 1],
    [1, 1, 1]
], dtype=np.float32)
dst = cv2.filter2D(img, -1, kernel)
plt.imshow(np.hstack((img, dst)), 'gray')
plt.show()
```
