```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('image.jpg')
# 图片形状
print(img.shape)
# 图片数据类型
print(img.dtype)
# 像素总数
print(img.size)
```

```python
face = img[100:500, 100:500]
plt.imshow(face[:,:,[2,1,0]])
plt.show()
```

```python
# 通道分割
b, g, r = cv2.split(img)
# 通道合并
img = cv2.merge((b, g, r))
plt.imshow(b, 'gray')
plt.show()
plt.imshow(g, 'gray')
plt.show()
plt.imshow(r, 'gray')
plt.show()
plt.imshow(img[:,:,[2,1,0]])
plt.show()
```

```python
# 通道分割
plt.imshow(img[:,:,0], 'gray')
plt.show()
plt.imshow(img[:,:,1], 'gray')
plt.show()
plt.imshow(img[:,:,2], 'gray')
plt.show()
```
