```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('sudoku.jpg')
edges = cv2.Canny(img, 38, 94)
plt.imshow(img[:,:,::-1])
plt.show()
plt.imshow(edges, 'gray')
plt.show()
```

```python
img = cv2.imread('sudoku.jpg', 0)
img1 = cv2.GaussianBlur(img, (5, 5), 1)
thresh = cv2.adaptiveThreshold(img1, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 17, 6)
plt.imshow(thresh, 'gray')
plt.show()
edges = cv2.Canny(thresh, 38, 94)
plt.imshow(edges, 'gray')
plt.show()
```

```python
img = cv2.imread('sudoku.jpg', 0)
def nothing(x):
    pass
cv2.namedWindow('image')
cv2.createTrackbar('maxVal', 'image', 0, 255, nothing)
cv2.createTrackbar('minVal', 'image', 0, 255, nothing)
while True:
    maxVal = cv2.getTrackbarPos('maxVal', 'image')
    minVal = cv2.getTrackbarPos('minVal', 'image')
    edges = cv2.Canny(img, minVal, maxVal)
    cv2.imshow('image', edges)
    if cv2.waitKey(1) == ord('q'):
        break
```
